```python
# 3) For each year, how many movies do not have a homepage?
movies ["release_date"] = pd.to_datetime (movies ["release_date"])

# aggiungere la variabile 'Year' al dataframe esistente
movies['Year'] = movies ["release_date"].dt.year
movies.Year.head()
```

```python
# 6) For each movie, compute the gross margin (difference between revenue and budget)
movies['difference'] = movies['revenue'] - movies['budget']
gross_margin = movies.groupby('movie_id').sum()[['difference']]
gross_margin.head()
```

```python
# 7) For each movie, compute the number of crew members
crew_memb = crew_normalized.groupby('movie_id').size().reset_index().rename(columns= {0:'num_crew_memb'})
crew_memb.head()
```

```python
# 9) For each language, compute the number of movies where such language is spoken.
number_movies = spoken_languages_normalized.groupby('iso_639_1').size().reset_index().rename(columns= {0:'num_movies_lang'})
number_movies.head()
```

```python
# 10)  For each company and each decade, compute the overall revenue
Year = movies.Year
def to_decade(year):
    return str(year)[:3] + '0'

movies['decade'] = movies['Year'].apply(to_decade)
movies.decade.head()
```
